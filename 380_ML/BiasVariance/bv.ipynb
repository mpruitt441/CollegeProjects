bias = $\mathbb{E}_{\bar{x}}[(\bar{g}(\bar{x})-f(\bar{x}))^{2}]$



```python
import numpy as np
import matplotlib.pyplot as plt

K = 100 # how many times we'll choose a random data set

B = []
for k in range(K):
    x1 = np.random.rand(1,2)
    b = np.mean(np.sin(x1))
    B.append(b)
print(x1)
gbar = np.mean(b)
xx = np.random.rand(1,K)*np.pi*2

bias = np.mean((gbar-np.sin(xx)))**2
print(bias)

A = []
for k in range(K):
    x1 = np.random.rand(1,2)*np.pi*2
    b= np.mean(np.sin(x1))
    A.append((b-gbar)**2)

var_x = np.mean(A)
xx = np.random.rand(1,K)*np.pi*2

var = var_x
print("variance = ", var)
print("bias = ", bias)
domain = np.linspace(0,2*np.pi)
plt.scatter(x1,[0,0])
plt.scatter(x1,np.sin(x1))
plt.plot(domain,np.zeros(len(domain)))
plt.plot(domain,np.sin(domain))
plt.plot(domain,b*np.ones(len(domain)))
plt.show()

### bias =
```
